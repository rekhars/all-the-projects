```python
corr=df.corr()
corr
```

```python
imp_fea=df.corr()['Creditability']
imp_fea
```

```python
imp_features=df.corr()['Creditability']>=0.02
imp_features
```

```python
print('Selected important features', imp_features[imp_features== True])
x = df[list(imp_features.index[imp_features])]
x.head()
```

```python
x=x.drop(['Creditability'],axis=1)
x.head()
```

```python
y = df['Creditability']
y.head()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.3, random_state=3)
x_train.head()
```

```python
gnb=GaussianNB()
gnb.fit(x_train, y_train)
```

```python
pred=gnb.predict(x_test)
pred
```

```python
gnb.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(pred, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pred, y_test)
```
